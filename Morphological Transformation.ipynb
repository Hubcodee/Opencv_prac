```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating blank image of 600x600
blank_img = np.zeros((600,600))
```

```python
plt.imshow(blank_img,cmap="gray")
```

```python
#Put text on blank img
img=cv2.putText(blank_img,text="ABCD",org=(50,300),fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=5,color=(255,255,255),thickness=25)
```

```python
plt.imshow(img,cmap="gray")
```

```python
kernel = np.ones((5,5),dtype=np.uint8)
```

```python
erode_img = cv2.erode(img,kernel,iterations=4)
```

```python
plt.imshow(erode_img,cmap="gray")
```

```python
dilute_img=cv2.dilate(img,kernel,iterations=4)
```

```python
plt.imshow(dilute_img,cmap="gray")
```

```python
dilute_eroded_img=cv2.dilate(erode_img,kernel,iterations=3)
```

```python
plt.imshow(dilute_eroded_img,cmap="gray")
```
