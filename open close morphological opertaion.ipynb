```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating blank image of 600x600
blank_img = np.zeros((600,600))
```

```python
plt.imshow(blank_img,cmap="gray")
```

```python
#Put text on blank img
img=cv2.putText(blank_img,text="ABCD",org=(50,300),fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=5,color=(255,255,255),thickness=25)
```

```python
plt.imshow(img,cmap="gray")
```

```python
kernel = np.ones((5,5),dtype=np.uint8)
```

```python
#Open operation
#Erosion + Dilute
open_img = cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)
```

```python
plt.imshow(open_img,cmap="gray")
```

```python
#Close operation
#Dilute + Erosion
close_img = cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel)
```

```python
plt.imshow(close_img,cmap="gray")
```
